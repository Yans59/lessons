### The Slice in Meridian plane of Magnetic dipole
refrence website:
[stremplot using matplotlib](https://matplotlib.org/stable/gallery/images_contours_and_fields/plot_streamplot.html#sphx-glr-gallery-images-contours-and-fields-plot-streamplot-py)

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# mask the region what we don't need
def mask(a,b,r,r1,r2):
    for i in range(len(r)):
        for j in range(len(r[i])):
            if r[i][j]>=r1:
                a[i][j]=np.nan
                b[i][j]=np.nan
            elif r[i][j]<=r2:
                a[i][j]=np.nan
                b[i][j]=np.nan
    return a,b
```

```python
M=-8e15 # T cm^-3

z,x=np.mgrid[-1:1.02:100j,-1:1.02:100j]
r=np.hypot(x,z) 
bx=3*M*x*z/r**5
bz=M*(3*z**2-r**2)/r**5
Bx,Bz=mask(bx,bz,r,1,0.1)

fig,ax=plt.subplots(figsize=(10,10))
ax.streamplot(x,z,bx,bz,density=[4,4],color=r,cmap='gray',linewidth=0.5)
ax.set_aspect('equal')
ax.set_xlim(-1.2,1.2)
ax.set_ylim(-1.2,1.2)
ax.set_title('Slice in Meridian plane of Magnetic dipole')

## save the figure if you need
# fig.savefig('dipole.pdf')
```
